<h1>Salary scale determination with PLR.<h1>

y = a + b1*x + b2*x^2 + b3*x^3 + b4*x^4 + .......+bN*x^N

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
#Load data
data_frame = pd.read_csv("salary.csv", sep = ";")
#Check data
data_frame
```

```python
#plot the dataset
plt.scatter(data_frame['level'], data_frame['salary'])
plt.xlabel('Experience (Year)')
plt.ylabel('Salary')
plt.show()
```

Let's compare Linear and Polynomial regressions and observe that PLR will produce a better result for this project!

```python
#Observe the Linear Regression model to this dataset
reg = LinearRegression()
reg.fit(data_frame[['level']], data_frame['salary'])

plt.xlabel('Experience (Year)')
plt.ylabel('Salary')

plt.scatter(data_frame['level'], data_frame['salary'])

x_axis = data_frame['level']
y_axis = reg.predict(data_frame[['level']])
plt.plot(x_axis, y_axis, color='green', label='Linear Regression')
plt.legend()
plt.show()
```

You can see that Linear regression is not suitable for this values of dataset

```python
#Degree = N (based on formula)
polynomial_regression = PolynomialFeatures(degree=2)
x_polynomial = polynomial_regression.fit_transform(data_frame[['level']])
```

```python
reg = LinearRegression()
reg.fit(x_polynomial, data_frame['salary'])
```

```python
y_head = reg.predict(x_polynomial)
plt.plot(data_frame['level'], y_head, color="red", label="Polynomial regression")
plt.legend()
plt.scatter(data_frame['level'], data_frame['salary'])
plt.show()
```

Now let's degree = 5 (N=5)

```python
#Degree = N (based on formula)
polynomial_regression = PolynomialFeatures(degree=5)
x_polynomial = polynomial_regression.fit_transform(data_frame[['level']])
reg = LinearRegression()
reg.fit(x_polynomial, data_frame['salary'])
y_head = reg.predict(x_polynomial)
plt.plot(data_frame['level'], y_head, color="red", label="Polynomial regression")
plt.legend()
plt.scatter(data_frame['level'], data_frame['salary'])
plt.show()
```
